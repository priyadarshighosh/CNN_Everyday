### Solution :

# Edge Detection

````python
edge_detection = np.array([
  [-1, -1, -1],
[-1,  8, -1],
[-1, -1, -1]
])                           # This is our edge Detection kernel
````

````python
edge_detection               # printing the Kernel
````

````python
image.shape                 # Shape of the kernel
````

We have to do it on grey scale image instead of colorful image as grey scale image has 1 layer so we can take 2 features like rows and columns

But for colorful image there are 3 layers so then we can put it in 3 features like rows1 , cols1 and layer

````python
k = edge_detection.shape[0]
k
````

````python
rows , cols = image.shape
rows , cols
````

Assigning Output matrix as 0

````python
output_matrix = np.zeros((rows - k + 1, cols - k + 1))
output_matrix
````

# Now we will Apply the loop

This is the function that is in-built in the CV2 but we have maunally written it to understand the function on a LOW LEVEL

````python
for i in range(1 , rows - 1):


    for j in range(1 ,cols -1):
        region =  image[i-1:i+2, j-1:j+2]

        output_matrix[i-1, j-1] = np.sum(region * edge_detection)
````

````python
output_matrix          # this is out output matrix
````

````python
cv2_imshow(output_matrix) # this is our output image
````

**Task**: Implement a function to perform image filtering and transformation that includes:

* **Filtering**: Apply a Gaussian blur to the image to reduce noise and smooth the image.
* **Transformation**: Apply a rotation to the image by a specified angle.


Hints:
* For Gaussian blur, you can create a Gaussian kernel and apply it using convolution. https://en.wikipedia.org/wiki/Gaussian_blur

### Solution :

Applying the Gaussian Blur

````python
gaussian_blur_kernel = (1/256) * np.array([
   [1,4,6,4,1],
   [4,16,24,16,4],                                 # we have applied the gaussian Blur to the applied image
   [6,24,36,24,6],
   [4,16,24,16,4],
   [1,4,6,4,1]])



output = cv2.filter2D(image, -1, gaussian_blur_kernel)

cv2_imshow(output)  # showing the output image
output.shape
````
